```julia
using Random
using LaserTag
using POMDPSimulators
using ParticleFilters
using POMDPPolicies
using POMDPModelTools
using POMDPs
```

```julia
rng = MersenneTwister(14)
m = gen_lasertag(rng=rng, robot_position_known=true)
p = FunctionPolicy(b->first(actions(m)))
f = SIRParticleFilter(m, 10000, rng=rng)
step = collect(stepthrough(m, p, f, max_steps=2, rng=rng))[1];
```

```julia
render(m, step)
```
